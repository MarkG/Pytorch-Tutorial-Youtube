```python
# groupby methods
# df.groupby('Origin').agg({'DepTime': ['min', 'max', 'mean'], 'DepDelay': ['min'], 'AirTime': ['sum']})
# groupby columns
df_agg = df.groupby('Origin').agg({'DepTime': ['min', 'max', 'mean'], 'DepDelay': ['min'], 'AirTime': ['sum']})
df_agg.columns.to_flat_index() # lists of column, agg tuples
df_agg.columns = ['_'.join(col) for col in df_agg.columns.to_flat_index()] # renames columns
df_agg.head()
```

```python
# new column
df['NewCol1'] = df['AirTime'] / 10
df = df.assign(NewCol2 = df['AirTime'].apply(lambda x: x / 10))
df.head()
```

```python
# sorting data
df.sort_values(['DepTime', 'Origin'], ascending=False).head()
```

```python
# Handling missing data
# df[['Airline', 'Origin']].isna().sum() # counts missing values
# df[['Airline', 'Origin']].dropna().head() # drops missing values
df[['Airline', 'Origin']].fillna('Unknown').head() # fills missing values
```

```python
# Combining Data
df1 = df.query('Origin == "EWR"').copy()
df2 = df.query('Origin == "JFK"').copy()

df_stack = pd.concat([df1, df2]) # stack DFs one on other

df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df_side = pd.concat([df1, df2], axis=1) # stack DFs side by side

df1.shape, df2.shape, df_stack.shape, df_side.shape
```

```python
# Merge data
# based on similar columns
df1 = df.groupby(['Airline','FlightDate'])[['DepDelay']].mean().reset_index()
df2 = df.groupby(['Airline','FlightDate'])[['ArrDelay']].mean().reset_index()
# merge with on when columns are same
pd.merge(df1, df2, on=['Airline','FlightDate'])
# or
df_merge = df1.merge(df2, on=['Airline','FlightDate'], suffixes=('_DepDelay', '_ArrDelay'))
df_merge.head()
# or with left_on and right_on when columns are different
df1.merge(df2, left_on=['Airline'], right_on=['Airline'])
```
